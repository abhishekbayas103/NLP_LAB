```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-hi"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
## Translation Function
```

```python
def translate_en_to_hi(text):
    tokens = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

text = "Public health information should be accessible to everyone."
print("English:", text)
print("Hindi:", translate_en_to_hi(text))
```

```python
## Translating Public Information Content
sentences = [
    "The government announced new health guidelines.",
    "Emergency services are available 24 hours a day.",
    "Education is essential for national development."
]

for s in sentences:
    print("\nEnglish:", s)
    print("Hindi:", translate_en_to_hi(s))
```
